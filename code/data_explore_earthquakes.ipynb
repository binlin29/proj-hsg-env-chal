```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random

# from mpl_toolkits.basemap import Basemap
import folium
```

```python
# load data
dataPath = "/mnt/data/"
df = pd.read_csv(dataPath+"earthquakes_consolidated_data.csv")
```

```python
df.head()
```

```python
# select close-to-CA data
# 32Â° 30' N to 42Â° N and 114Â° 8' W to 124Â° 24' W
df = df.dropna(subset=['place'])
df_CA = df[df['place'].str.contains("CA")].reset_index()

# df_CA = df[ 
#   (32 < df['latitude'])  & (df['latitude'] < 42) &
#   (-115 > df['longitude']) & (df['longitude'] >  -125) ].reset_index()
# df_CA.head()
```

```python
df_CA
```

```python
num_eqCA = len(df_CA) 
print("#records in CA: {}".format(num_eqCA))
```

```python
# plot 1,000 randomly selected earthquakes

m = folium.Map(location=(37, -120), zoom_start=5, width='50%', length='50%')

for i in random.sample(range(0, len(df_CA)), 1000):
    if df_CA.iloc[i]['mag']>3:
        folium.Circle(
            location=[df_CA.iloc[i]['latitude'], df_CA.iloc[i]['longitude']],weight=1,
            radius=df_CA.iloc[i]['mag']*10000, color='red', opacity=0.3,fill_color='red',fill_opacity=0.1,
        ).add_to(m)
    else:
        folium.Circle(
            location=[df_CA.iloc[i]['latitude'], df_CA.iloc[i]['longitude']],weight=1,
            radius=5, color='red', opacity=0.3,fill_color='red',fill_opacity=0.1,
        ).add_to(m)




# # Same as before, we save it to file
# m.save('earthquakes.html')
```
